  <span style="background-color: #FFD700">**Summarize the terms below:**</span>

 <span style="background-color: #AFEEEE">**1. Feature selection**</span>

 Possible answer: The process of choosing which features in the dataset to keep/discard before training a model. Often involves domain-specific knowledge and statistical analysis (eg. analysing correlation between features).

<span style="background-color: #AFEEEE">**2. Hold out data**</span>

Possible answer: Part of the dataset that is reserved for testing; not used in model training. Model has never seen this data until testing, and this data will not be used to alter the model weights

<span style="background-color: #AFEEEE">**3. Parameter**</span>

Possible answer: Synonym for weights; the trainable part of a model that affects the output. Similar idea to coefficients in an equation; adjust the coefficient and you get a different result, for the same input x.


 <span style="background-color: #AFEEEE">**4. Epoch**</span>

 Possible answer: 1 epoch is completed during training when the model has seen all points in the training set once. For example, if the training set has 10 samples, after 5 epochs the model will have seen all 10 samples 5 times.


 <span style="background-color: #AFEEEE">**5. Inference**</span>

Possible answer: Using a machine learning model to generate a prediction based on new data. Usually is referring to when the model has been trained and is being used/deployed in the real world, and is fed new "real world" data to get predictions
